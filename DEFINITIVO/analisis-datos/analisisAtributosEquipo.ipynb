```python
# Función para calcular diferencias de posiciones
def calcular_diferencias(df_original, df_ordenado, atributo):
    orden_original = df_original['nombre'].tolist()
    orden_nuevo = df_ordenado['nombre'].tolist()

    for idx, nombre in enumerate(orden_original):
        pos_original = idx
        pos_nueva = orden_nuevo.index(nombre)
        diferencias[nombre][atributo] = (pos_original-pos_nueva)
```

```python
for atributo in lista_atributos:
  df_sorted_by_atributo1 = df.sort_values(by=atributo, ascending=False).reset_index(drop=True)
  calcular_diferencias(df_sorted_by_main, df_sorted_by_atributo1, atributo)
```

```python
# Imprimir las diferencias de posición para cada nombre
for nombre, difs in diferencias.items():
    print(f"{nombre}: {difs}")
```

```python
# Convertir el diccionario de diferencias a un DataFrame
df_diferencias = pd.DataFrame.from_dict(diferencias, orient='index')
```

```python
# Restablecer el índice para que los nombres sean una columna en lugar del índice
df_diferencias.reset_index(inplace=True)
df_diferencias.rename(columns={'index': 'nombre'}, inplace=True)

# Imprimir el DataFrame de diferencias
df_diferencias
```
